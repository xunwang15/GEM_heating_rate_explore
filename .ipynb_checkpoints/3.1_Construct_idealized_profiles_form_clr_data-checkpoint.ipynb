## Description

This notebook shows how I constructed the idealized profiles for rrtmg. This includes the base profiles (no convection) and the perturbed profiles.

For the base profiles, I generated three types:
1. No convectvie perturbation in the troposphere
2. convectvie perturbation at the edge of the convection
3. convective perturbation at the center of the convection

For the perturbed profiles, I generated two files:
1. perturbations in overworld and lowermost altitude ranges, respectively
2. perturbations layer by layer between ~140 hPa and ~88 hPa

There is no need to rerun this notebook unless you need to generate new idealized dataset. Jump to "Store the data" section at the bottom to read codes showing how I stored the data.

```python

def get_dist(lat1,lat2,lon1,lon2):
    dlat = abs(lat1-lat2)
    dlon = abs(lon1-lon2)
    R = 6371
    a = np.sin(np.radians(dlat)/2)**2+np.cos(np.radians(lat1))*np.cos(np.radians(lat2))*np.sin(np.radians(dlon)/2)**2
    c = 2*math.atan2(np.sqrt(a),np.sqrt(1-a))
    return R*c

def gavg(data,lat):
    """calculate global average
    e.g., x1=gavg(d1['t2m'])"""
    datam = np.nanmean(data,axis=-1)
    wgt1 = np.cos(np.deg2rad(lat))
    wgt = wgt1[np.newaxis,:]*(datam*0+1)
    ga = np.nansum(wgt*datam,axis=-1)/np.nansum(wgt,axis=-1)
    return ga

def gavg2(data,lat):
    """calculate global average
    e.g., x1=gavg(d1['t2m'])"""

    wgt1 = np.cos(np.deg2rad(lat))
    wgt = wgt1[np.newaxis,:]*(data*0+1)
    ga = np.nansum(wgt*data,axis=-1)/np.nansum(wgt,axis=-1)
    return ga

def sat_mix_mk(p,t):
    idx1 = np.where(t>273.5)
    esi = np.zeros_like(t)
    esi[idx1] = np.exp(54.842763-6763.22/t[idx1]-4.210*np.log(t[idx1]) +\
        0.000367*t[idx1]+np.tanh(0.0415*(t[idx1]-218.8))*(53.878-1331.22/t[idx1] - \
        9.44523*np.log(t[idx1])+0.014025*t[idx1]))/ 100. # hPa
    idx2 = np.where(t<=273.5)
    esi[idx2] = np.exp(9.550426-5723.265/t[idx2]+3.53068*np.log(t[idx2])-0.00728332*t[idx2])/100.0 # hPa
    h2o=esi/(p-esi) # the mixing ratio

    idx3 = np.where(esi*100>p)
    if np.shape(idx3)[1]>0:
        h2o[idx3]=esi[idx3]/p[idx3] # if esi is too large, use specific humidity
    idx4 = np.where(h2o>=1)
    if np.shape(idx4)[1]>0:
        h2o[idx4] = np.nan
    return h2o*1e6
```

### Scheme 2: use a convective center and radius around it

#### This is the chosen scheme

#### First compute the distance of all grid points to the convective center

```python
dist_mtx = np.zeros((nlat,nlon))
for i in range(nlat):
    for j in range(nlon):
        dist_mtx[i,j]=get_dist(latc,lat[i],lonc,lon[j])
```

```python
nlev = 77 #77
ndist = 100
iwc_profile_pos = np.zeros((nlev,ndist))
iwc_profile_neg = np.zeros((nlev,ndist))

t_profile_pos = np.zeros((nlev,ndist))
t_profile_neg = np.zeros((nlev,ndist))

wv_profile_pos = np.zeros((nlev,ndist))
wv_profile_neg = np.zeros((nlev,ndist))

ta_profile_pos = np.zeros((nlev,ndist))
ta_profile_neg = np.zeros((nlev,ndist))

wva_profile_pos = np.zeros((nlev,ndist))
wva_profile_neg = np.zeros((nlev,ndist))
trop_pos = np.zeros((ndist))
trop_neg = np.zeros((ndist))
dist_range = np.arange(10,1010,10)
```

```python
ts_pos = np.zeros((ndist))
ts_neg = np.zeros((ndist))
ps_pos = np.zeros((ndist))
ps_neg = np.zeros((ndist))
```

#### Compute the average water vapor and temperature profiles in each radius range

```python
for i in range(ndist):
    if i>=1:
        idx = np.where(np.logical_and(dist_mtx>dist_range[i-1],dist_mtx<=dist_range[i]))
    else:
        idx = np.where(dist_mtx<=dist_range[i])
    ii1 = np.where(lon[idx[1]]>=lonc)[0]
    ii2 = np.where(lon[idx[1]]<=lonc)[0]
    iwc_profile_pos[:,i] = np.nanmean(iwc_grid[:,idx[0][ii1],idx[1][ii1]],axis=-1)
    iwc_profile_neg[:,i] = np.nanmean(iwc_grid[:,idx[0][ii2],idx[1][ii2]],axis=-1)
    t_profile_pos[:,i] = np.nanmean(tt_grid[:,idx[0][ii1],idx[1][ii1]],axis=-1)
    t_profile_neg[:,i] = np.nanmean(tt_grid[:,idx[0][ii2],idx[1][ii2]],axis=-1)
    wv_profile_pos[:,i] = np.nanmean(hu_grid[:,idx[0][ii1],idx[1][ii1]],axis=-1)
    wv_profile_neg[:,i] = np.nanmean(hu_grid[:,idx[0][ii2],idx[1][ii2]],axis=-1)
    
    ta_profile_pos[:,i] = np.nanmean(tt_anom[:,idx[0][ii1],idx[1][ii1]],axis=-1)
    ta_profile_neg[:,i] = np.nanmean(tt_anom[:,idx[0][ii2],idx[1][ii2]],axis=-1)
    wva_profile_pos[:,i] = np.nanmean(hu_anom[:,idx[0][ii1],idx[1][ii1]],axis=-1)
    wva_profile_neg[:,i] = np.nanmean(hu_anom[:,idx[0][ii2],idx[1][ii2]],axis=-1)

    trop_pos[i] = np.nanmean(tropp[idx[0][ii1],idx[1][ii1]])
    trop_neg[i] = np.nanmean(tropp[idx[0][ii2],idx[1][ii2]])
```

```python
for i in range(ndist):
    if i>=1:
        idx = np.where(np.logical_and(dist_mtx>dist_range[i-1],dist_mtx<=dist_range[i]))
    else:
        idx = np.where(dist_mtx<=dist_range[i])
    ii1 = np.where(lon[idx[1]]>=lonc)[0]
    ii2 = np.where(lon[idx[1]]<=lonc)[0]

    ts_pos[i] = np.nanmean(ts[idx[0][ii1],idx[1][ii1]])
    ts_neg[i] = np.nanmean(ts[idx[0][ii2],idx[1][ii2]])
    
    ps_pos[i] = np.nanmean(ps[idx[0][ii1],idx[1][ii1]])
    ps_neg[i] = np.nanmean(ps[idx[0][ii2],idx[1][ii2]])
```

```python
t_profile_all = np.zeros((nlev,2*ndist))
t_profile_all[:,:100]=t_profile_neg[:,::-1]*1.
t_profile_all[:,100:] = t_profile_pos*1.
wv_profile_all = np.zeros((nlev,2*ndist))
wv_profile_all[:,:100]=wv_profile_neg[:,::-1]*1.
wv_profile_all[:,100:] = wv_profile_pos*1.
iwc_profile_all = np.zeros((nlev,2*ndist))
iwc_profile_all[:,:100]=iwc_profile_neg[:,::-1]*1.
iwc_profile_all[:,100:] = iwc_profile_pos*1.

ta_profile_all = np.zeros((nlev,2*ndist))
ta_profile_all[:,:100]=ta_profile_neg[:,::-1]*1.
ta_profile_all[:,100:] = ta_profile_pos*1.
wva_profile_all = np.zeros((nlev,2*ndist))
wva_profile_all[:,:100]=wva_profile_neg[:,::-1]*1.
wva_profile_all[:,100:] = wva_profile_pos*1.
trop_all = np.zeros((2*ndist))
trop_all[:100]=trop_neg[::-1]*1.
trop_all[100:] = trop_pos*1.

dist_range_all = np.arange(-1000,1010,10)
#dist_range_all = np.arange(-500,505,5)
dist_range_all = (dist_range_all[1:]+dist_range_all[:-1])/2.
```
