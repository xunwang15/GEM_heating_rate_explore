```python
def sat_mix_mk(p,t):
    idx1 = np.where(t>273.5)
    if t>273.5:

        esi = np.exp(54.842763-6763.22/t-4.210*np.log(t) +\
        0.000367*t+np.tanh(0.0415*(t-218.8))*(53.878-1331.22/t - \
        9.44523*np.log(t)+0.014025*t))/ 100. # hPa
    if t<=273.5:

        esi = np.exp(9.550426-5723.265/t+3.53068*np.log(t)-0.00728332*t)/100.0 # hPa
    h2o=esi/(p-esi) # the mixing ratio
    if esi*100>p:
        h2o=esi/p
    if h2o>=1:
        h2o = np.nan
    return h2o*1e6
```

#### Get GEM data at 0001 min, the beginning of the model run (before convection), to compute the background mean for GEM

```python
indir = '/storage/xwang/transfer_data/data/GEM_1p0_griddata/'
ffw0=xr.open_dataset(indir+'2013082512_0001m_1p0_h2o_mix_gz.nc')
ff0=xr.open_dataset(indir+'2013082512_0001m_1p0_tt_theta_thetav_gz.nc')
lat1 = 40;lat2=52
wv0=ffw0.h2o_mix.sel(lat=slice(lat1,lat2)).data
tt0=ff0.tt.sel(lat=slice(lat1,lat2)).data
alt = ff0.level.data

#lat1 = 45;lat2=52
lat = ff0.lat.sel(lat=slice(lat1,lat2)).data
lon=ff0.lon.data
nlat = np.size(lat)
nlon=np.size(lon)

iilev1 = 32 
iilev2 =37
print(alt[iilev1:iilev2+1],'km')
wvm0 = np.nanmean(wv0[iilev1:iilev2,:],axis=-1)
wvstd0 = np.nanstd(wv0[iilev1:iilev2,:],axis=-1)
ttm0 = np.nanmean(tt0[iilev1:iilev2,:],axis=-1)
```

#### Get Clausius Clapeyron curve

```python
t_cc = np.arange(-20,5,1)
sat_cc = np.zeros_like(t_cc)
ttm0 = np.nanmean(tt0[iilev1:iilev2,:])
wvm0 = np.nanmean(wv0[iilev1:iilev2,:])
for i in range(len(t_cc)):
    sat_cc[i] = sat_mix_mk(100,t_cc[i]+ttm0)-wvm0
```

### 2. Get rrtmg Heating Rate data

The rrtmg is run by feeding the profiles stored in: idealized_gem_profiles_p_t_wv_scatter_perturb_base.nc

This dataset generation and storage is recorded in 2_GEM_vs_SEACR4S_ER-2_data.ipynb

```python
indir1 = '/storage/xwang/transfer_data/data/heating_rate_result_data/rrtmg_retrieved_data/'
f1 = xr.open_dataset(indir1+'idealized_gem_profiles_p_t_wv_scatter_perturb_base_both_up_dw_comp_clr_rrtmg_out.nc')

rlw_hr = f1.lw_hr.data
rsw_hr = f1.sw_hr.data
rpr = f1.lw_pr.data[0,0,:]
nset = len(f1.n2.data)-1
f1.close()

### Get delta heating rate
ilev=23
rdel_lw_hr = rlw_hr[:,1:,ilev]-rlw_hr[:,0,ilev][:,np.newaxis]
rdel_sw_hr = rsw_hr[:,1:,ilev]-rsw_hr[:,0,ilev][:,np.newaxis]
```

```python
### read the wv and T perturbations added to the base profiles. 
### The wv and T perturbations can be used to map the delta heating rate distribution into the del_T vs del_wv domain
indir2 = '/storage/xwang/transfer_data/data/heating_rate_result_data/'
f2 = xr.open_dataset(indir2+'idealized_gem_profiles_p_t_wv_scatter_perturb_base.nc')
ilevp = 54 # perturb level
del_t = f2.T_ave.data[0,1:,ilevp]-f2.T_ave.data[0,0,ilevp]
del_wv = (f2.gas_mr_wv.data[0,1:,ilevp]-f2.gas_mr_wv.data[0,0,ilevp])*1e6 # convert to ppmv
del_wv = np.array([round(i,0) for i in del_wv])


### Map the delta heating rate distribution into the del_T vs del_wv domain
wvbin = np.arange(-10,51,1) # water vapor anomaly bin
ttbin = np.arange(-20,21,1) # temperature anomaly bin
nwvbin = np.size(wvbin)-1
nttbin = np.size(ttbin)-1
rlw_hr_mtx = np.zeros((3,nwvbin,nttbin))
rsw_hr_mtx = np.zeros((3,nwvbin,nttbin))
for i in range(nset):
    ii1 = np.where(wvbin==del_wv[i])[0][0]
    ii2 = np.where(ttbin==del_t[i])[0][0]
    rlw_hr_mtx[:,ii1,ii2]=rdel_lw_hr[:,i]
    rsw_hr_mtx[:,ii1,ii2]=rdel_sw_hr[:,i]
```

### 3. $\Delta{HR}=\frac{\delta{HR}}{\delta{T}}*\Delta{T}+\frac{\delta{HR}}{\delta{H2O}}*\Delta{H2O}$

Show the distribution of the computed $\Delta{HR}$

#### First read the kernel data

```python
indir = '/storage/xwang/transfer_data/data/heating_rate_result_data/'
ff = xr.open_dataset(indir+'heating_rate_kernels_ovw_lms_t_wv_140-88hPa.nc')
```

```python
### for this notebook, I only used OVW
i0 = 0 # 0 = ovw, 1 = lms
### since we are only showing results on 100 hPa, we only read kernels for 100 hPa
ik = 10 
t_kn_lw = ff.t_kn_lw.data[i0,ik,:]
wv_kn_lw = ff.wv_kn_lw.data[i0,ik,:]
t_kn_sw = ff.t_kn_sw.data[i0,ik,:]
wv_kn_sw = ff.wv_kn_sw.data[i0,ik,:]
```

#### Also read the perturbation profiles

```python
indir2 = '/storage/xwang/transfer_data/data/heating_rate_result_data/'
f2 = xr.open_dataset(indir2+'idealized_gem_profiles_p_t_wv_scatter_perturb_base.nc')
delta_t = f2.T_ave.data[0,1:,:]-f2.T_ave.data[0,0,:]
delta_wv = (f2.gas_mr_wv.data[0,1:,:]-f2.gas_mr_wv.data[0,0,:])*1e6
for i in range(nset):
    delta_wv[i,ilevp]=round(delta_wv[i,ilevp])
```

#### get the dot product


```python

hr_lw_t = np.zeros((nset))
hr_sw_t = np.zeros((nset))
hr_lw_wv = np.zeros((nset))
hr_sw_wv = np.zeros((nset))

ilevp=54
for j in range(nset):
    hr_lw_t[j] = np.nansum(t_kn_lw*delta_t[j,:])
    hr_sw_t[j] = np.nansum(t_kn_sw*delta_t[j,:])
    hr_lw_wv[j] = np.nansum(wv_kn_lw*delta_wv[j,:])
    hr_sw_wv[j] = np.nansum(wv_kn_sw*delta_wv[j,:])
hr_lw_both = hr_lw_t+hr_lw_wv
hr_sw_both = hr_sw_t+hr_sw_wv
```

#### Map the $\Delta{HR}$ distribution into the del_T vs del_wv domain

```python
hr_lw_mtx = np.zeros((3,nwvbin,nttbin))
hr_sw_mtx = np.zeros((3,nwvbin,nttbin))
for i in range(nset):
    ii1 = np.where(wvbin==del_wv[i])[0][0]
    ii2 = np.where(ttbin==del_t[i])[0][0]
    hr_lw_mtx[:,ii1,ii2]=hr_lw_both[i]
    hr_sw_mtx[:,ii1,ii2]=hr_sw_both[i]
```

#### Plot the $\Delta{HR}$ distribution

```python
fig,ax = plt.subplots(nrows=1,ncols=3)
fig.set_size_inches(20,5, forward=True)
colors=['b','orange','g','magenta','cyan']
titles = ['LW','SW','LW+SW']

### LW heating rate
levels=np.arange(-4.5,4.6,0.5)
var = hr_lw_mtx[0,:]*1.
#var[var==0]=np.nan
image=ax[0].contourf(ttx,wvy,var,cmap=cm_bgyr,levels=levels,extend='both')
cbaxes = fig.add_axes([0.91,0.15,0.01,0.7])                             # setup colorbar
cc = fig.colorbar(image, ax=ax[0],cax=cbaxes)  
cc.set_ticks(levels[::3])
clabels = np.array([str(i) for i in levels[::3]])
cc.set_ticklabels(clabels)
cc.ax.tick_params(labelsize=13) 
cc.set_label('K/day',fontsize=14) 

### SW heating rate
levels=np.arange(-4.5,4.6,0.1)
var = hr_sw_mtx[0,:]*1.
image=ax[1].contourf(ttx,wvy,var,cmap=cm_bgyr,levels=levels,extend='both')

### LW+SW heating rate
levels=np.arange(-4.5,4.6,0.5)
var = hr_sw_mtx[0,:]+hr_lw_mtx[0,:]
image=ax[2].contourf(ttx,wvy,var,cmap=cm_bgyr,levels=levels,extend='both')


level1 = np.sort(twmoist.flatten())[::-1][7]
level2 = np.sort(twmoist.flatten())[::-1][10]
level3 = np.sort(twmoist.flatten())[::-1][-1]
perc = np.sort(twmoist.flatten())[::-1]/np.nansum(twmoist)
kk=0
while(sum(perc[:kk])<=0.8):
    kk+=1
print(np.sort(twmoist.flatten())[::-1][kk],level1,np.nansum(np.sort(twmoist.flatten())[::-1][:7])/np.nansum(twmoist))

temp= np.sort(np.array(list(set(twmoist.flatten()))))[1:]
temp = temp[::30]*1.
levels = temp*1.
for i in range(3):
    ax[i].contour(ttx,wvy,twmoist,colors='r',linewidths=2,levels=[level2])
    ax[i].contour(ttx,wvy,twmoist,colors='r',linewidths=1,levels=[level3])
    ax[i].plot(t_cc,sat_cc)
    ax[i].plot(tt90,tt90*slp[0]+slp[1],color='k',linewidth=2)

    ax[i].set_xlabel('delta T(K)',fontsize=12)
    ax[i].set_ylabel('delta H2O(ppmv)',fontsize=12)
    ax[i].set_xlim([-20,10])
    ax[i].set_ylim([-10,50])
    ax[i].set_title(titles[i],fontsize=15)
```

## Conlcusion

1. The del_HR gradient is dominated by the temperature gradient.

2. For the 90% distribution, the del_HR is positive. (LW del_HR Range: 1.14 K/day ~ 2.31 K/day; SW del_HR range: 0.015 K/day ~ 0.067 K/day)

3. As del_WV increase, the kernel result tends to overestimate the LW radiative cooling and SW radiative heating due to water vapor.

3. This means the equation: $\Delta{HR}=\frac{\delta{HR}}{\delta{T}}*\Delta{T}+\frac{\delta{HR}}{\delta{H2O}}*\Delta{H2O}$ may not be accurate. Try quadratic or logarithm
