#### For better comparison, interpolate the rrtmg flux data onto GEM pressure levels

```python
#nnlat = 10
nnlat = 100
uflux_lw2 = np.zeros((62,nnlat,1024))
dflux_lw2 = np.zeros((62,nnlat,1024))
uflux_sw2 = np.zeros((62,nnlat,1024))
dflux_sw2 = np.zeros((62,nnlat,1024))
for i in range(nnlat):
    for j in range(1024):
        fft = interpolate.interp1d(np.log(pr_lw[:,i,j]),uflux_lw[:,i,j],fill_value='extrapolate')
        uflux_lw2[:,i,j] = fft(np.log(prgem[:,i,j]))
        fft = interpolate.interp1d(np.log(pr_lw[:,i,j]),dflux_lw[:,i,j],fill_value='extrapolate')
        dflux_lw2[:,i,j] = fft(np.log(prgem[:,i,j]))
        
        fft = interpolate.interp1d(np.log(pr_sw[:,i,j]),uflux_sw[:,i,j],fill_value='extrapolate')
        uflux_sw2[:,i,j] = fft(np.log(prgem[:,i,j]))
        fft = interpolate.interp1d(np.log(pr_sw[:,i,j]),dflux_sw[:,i,j],fill_value='extrapolate')
        dflux_sw2[:,i,j] = fft(np.log(prgem[:,i,j]))
```

### 2.3 The discrepancies could be due to:
The GEM radiatvie transfer calculation is different from rrtmg.
The rrtmg does not take into account of precipitation. There is snow, hail, etc. in the GEM.

#### I notice the bias could be from locations where there is large effective radius but no iwc or lwc.

#### To test that hypothesis, let's remove data with large effective radius but no iwc and see if the bias gets smaller.

```python
clm = ff.cloud_mask_3d.isel(lat=slice(latrange[0],latrange[1])).data
iwc = ff.water_content_ice.isel(lat=slice(latrange[0],latrange[1])).data
iwc+=ff.implicit_cloud_solid_water_content.isel(y=slice(latrange[0],latrange[1])).data
lwc = ff.water_content_rain.isel(lat=slice(latrange[0],latrange[1])).data
lwc+=ff.implicit_cloud_liquid_water_content.isel(y=slice(latrange[0],latrange[1])).data
cwc = ff.water_content_cloud.isel(lat=slice(latrange[0],latrange[1])).data
```

```python
### Find location with large efri but no iwc
flag1 = np.ones((100,1024)) # a flag of ones
for i in range(100):
    for j in range(1024):
        iii=np.where(np.logical_and(efri[:,i,j]>0,iwc[:,i,j]==0))[0]
        if np.shape(iii)[0]>0:
            flag1[i,j]=0 # for these locations set the flag to be zero
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)
fig.set_size_inches(13,5, forward=True)
# heating rate data*flag -> we removed the data with large effective radius but no iwc.
lwhr2 = lwhr*flag1[np.newaxis,:] 
hr_lw2 = hr_lw*flag1[np.newaxis,:]
swhr2 = swhr*flag1[np.newaxis,:]
hr_sw2 = hr_sw*flag1[np.newaxis,:]

ax[0].scatter(lwhr2[1:-1,:].flatten(),hr_lw2[1:-1,:].flatten(),s=1)
ax[0].plot(np.arange(-40,40,1),np.arange(-40,40,1),color='k')

ax[1].scatter(swhr2[1:-1,:].flatten(),hr_sw2[1:-1,:].flatten(),s=1)
ax[1].plot(np.arange(0,25,1),np.arange(0,25,1),color='k')
ax[0].set_xlabel('GEM LW HR (K/day)',fontsize=12)
ax[0].set_ylabel('RRTMG LW HR (K/day)',fontsize=12)
ax[1].set_xlabel('GEM SW HR (K/day)',fontsize=12)
ax[1].set_ylabel('RRTMG SW HR (K/day)',fontsize=12)
```
